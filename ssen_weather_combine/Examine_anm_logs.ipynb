```r
getwd()
setwd('./College/ssen/ssen_weather_combine')
data <- read.csv("./database/combined_ssen_weather.csv", header=TRUE, sep=",", dec=".")

data$dt <- as.POSIXct(data$elec_timestamp, origin="1970-01-01",tz="GMT")

# data$dt <- as.POSIXct(data$timestamp,format="%Y-%m-%d %H:%M:%S",tz="GMT")

plot(data$dt, data$approxWindSpeed, main= "windspeed", type= "l", col = "blue")
plot(data$approxWindSpeed, data$anm_generated, main='windspeed vs generation', type="p", pch='.')
plot(data$approxWindSpeed, data$non_anm_generated, main='windspeed vs generation', type="p", pch='.')

unique(data$operation_core)
levels(data$operation_core)
levels(factor(data$operation_core))
data$operation_core <-factor(data$operation_core)
       
data$operation_core[1]
 # core, 1, 1A, 2, 2A, 2B, 3, 4, 4A

anmlabels <- c(0,1,4)
names(anmlabels) <- c("ok", "warning", "remove")

anmlabels
anmlabels["ok"]

anmlabels[data$operation_core[1]]

plot(data$approxWindSpeed, anmlabels[data$operation_2B], main='windspeed vs removals', type="p")
levels(data$operation_core) <- list("0"="ok","1"="warning","2"="remove")
levels(data$eqpt_core) <- list("0"="ok","1"="warning","2"="remove")

data$anmcount <- anmlabels[data$operation_core]+
  anmlabels[data$operation_1]+
  anmlabels[data$operation_1A]+
  anmlabels[data$operation_2]+
  anmlabels[data$operation_2A]+
  anmlabels[data$operation_2B]+
  anmlabels[data$operation_3]+
  anmlabels[data$operation_4]+
  anmlabels[data$operation_4A]

plot(data$approxWindSpeed, data$anmcount, main='windspeed vs removals', type="p")
plot(data$demand, data$anmcount, main='windspeed vs removals', type="p")
plot(data$anm_generated, data$anmcount, main='windspeed vs removals', type="p")
plot(data$non_anm_generated, data$anmcount, main='windspeed vs removals', type="p")

table(data$eqpt_core)
table(data$operation_core)

as.numeric(data$operation_core[1])
as.numeric(as.character(data$operation_core[1]))

hist(data$approxWindSpeed, 60,main= "windspeed", col = "blue")
hist(data$approxWindSpeed[data$operation_2B=="remove"], 60,main= "number of shutdowns vs windspeed", col = "blue")
hist(data$approxWindSpeed[data$operation_core=="remove"], 60,main= "number of shutdowns vs windspeed", col = "blue")
hist(data$approxWindSpeed[data$operation_2=="remove"], 60,main= "number of shutdowns vs windspeed", col = "blue")
hist(data$approxWindSpeed[data$operation_2A=="remove"], 60,main= "number of shutdowns vs windspeed", col = "blue")
hist(data$approxWindSpeed[data$operation_1A=="remove"], 60,main= "number of shutdowns vs windspeed", col = "blue")
```
